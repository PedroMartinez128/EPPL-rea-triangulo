## Formula para sacar el area de un triangulo

$A = (B * H)/2$

---
## Datos de entrada
* $B$ ?
* $H$ ?

```python
# Calcular el area de un triangulo
# Variables A, B, H
# Formula A = (B * H) / 2
# Capturar valor de B, H
print("Calcular el area de un triangulo")
B = float(input("Ingrese el valor de la base: "))
H = float(input("Ingrese el valor de la altura: "))
A = (B * H) / 2
print("El area del triangulo es: ", A)
print("Pedro Martinez NL = 34")
```
